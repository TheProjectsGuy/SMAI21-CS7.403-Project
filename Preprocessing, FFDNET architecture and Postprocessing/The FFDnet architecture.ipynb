```python
#import statements
import numpy as np
import tensorflow as tf
from keras.layers import Input, Flatten, Dense, Conv2D
from tensorflow.keras.layers import Input,MaxPool2D,Conv2D,UpSampling2D,Activation,BatchNormalization,Subtract
from keras.models import Model
#from keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.utils import plot_model
```

```python
#This method creates the architecture for FFDnet as specified in the paper for training with colored images
def model_design_for_colored_images():
    input = Input(shape=(50,50,3),name='Input Layer')
    x = Conv2D(96,kernel_size= (3,3), padding='same',name='Convolution_Layer_1')(input)
    x = Activation('relu',name='Activation_Layer_1')(x)
    for i in range(10): # as for color the FFDnet has a total of 12 layers
        x = Conv2D(96, kernel_size=(3,3), padding='same',name='Convolution_Layer'+str(i+2))(x)
        x = BatchNormalization(axis=-1,name='Batch_Normalization_Layer'+str(i+1))(x)
        x = Activation('relu',name='Activation_Layer'+str(i+2))(x)   
    x = Conv2D(3, kernel_size=(3,3), padding='same',name='Convolution_Layer_12')(x)
    x = Subtract(name='noise_removal')([input, x])   
    model = Model(input,x)
    model.compile(optimizer=tf.keras.optimizers.Adam(1e-03), loss=tf.keras.losses.MeanSquaredError())
    print(model.summary())
    
```

```python
#viewing the architecture for
model_design_for_colored_images()
```

```python
#This method creates the architecture for FFDnet as specified in the paper for training with grayscale images
def model_design_for_grayscale_images():
    input = Input(shape=(70,70,1),name='Input Layer')
    x = Conv2D(64,kernel_size= (3,3), padding='same',name='Convolution_Layer_1')(input)
    x = Activation('relu',name='Activation_Layer_1')(x)
    for i in range(13): # as for grayscale the FFDnet has a total of 15 layers
        x = Conv2D(64, kernel_size=(3,3), padding='same',name='Convolution_Layer'+str(i+2))(x)
        x = BatchNormalization(axis=-1,name='Batch_Normalization_Layer'+str(i+1))(x)
        x = Activation('relu',name='Activation_Layer'+str(i+2))(x)   
    x = Conv2D(3, kernel_size=(3,3), padding='same',name='Convolution_Layer_12')(x)
    x = Subtract(name='noise_removal')([input, x])   
    model = Model(input,x)
    model.compile(optimizer=tf.keras.optimizers.Adam(1e-03), loss=tf.keras.losses.MeanSquaredError())
    print(model.summary())
    
```

```python
# this method creates the patches for the given image
def generate_patches(filepath,patch_size,crop_dim,save_dir):
    image = cv2.imread(filepath) 
    #print(filepath)
    #print(image)
    filename=filepath.split('/')[-1].split('.')[0]
    #print("filename is ",filename)
    height, width , channels= image.shape
    patches = []

    for crop_size in crop_dim: #We will crop the image to different sizes
        crop_h, crop_w = int(height*crop_size),int(width*crop_size)
        image_scaled = cv2.resize(image, (crop_w,crop_h), interpolation=cv2.INTER_CUBIC)
        for i in range(0, crop_h-patch_size+1, patch_size):
            for j in range(0, crop_w-patch_size+1, patch_size):
                x = image_scaled[i:i+patch_size, j:j+patch_size] # This gets the patch from the original image with size patch_size x patch_size
                print("printing patch ")
                print(x)
                
                if save_dir!=None:  #If save_dir is specified we will save the patches to the given directory
                    print("within save dir condition")
                    if not os.path.exists(save_dir):
                        os.mkdir(save_dir)
                    filepath_for_generated_patch=save_dir+'/'+filename+'_'+str(crop_h)+'_'+str(i)+'_'+str(j)+'.jpg'# Creating a distinct patch filename
                    cv2.imwrite(filepath_for_generated_patch,x) # Saving the  image to the path
            patches.append(x)
    #print("The final value of patches set is shape is", patches.shape)
    #print(patches)
    return patches

```

```python
def data_creator(data_dir,save_dir):
    '''This function create data for all the images present in a directory'''
    files_list=os.listdir(data_dir)
    #files_list_2=files_list[0:10] # this is only for testing purpose_it has_to_be_removed_in_the_final_version
    #print(save_dir)  
    patch_size = 40 # As of now going with patch size of 40 later shall change it to 70 to meet the paper  
    crop_dim = [1, 0.8, 0.7,0.5]
    data=[]
  
    for file in files_list: 
        if file.endswith('.jpg'):
            patches=generate_patches(data_dir+'/'+file,patch_size,crop_dim,save_dir)
            data.append(patches)

    return data
```
